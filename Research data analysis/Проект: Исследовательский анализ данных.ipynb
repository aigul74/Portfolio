```python
def days_exposition_price(x):
    try:
        if x['last_price'] < 4500000:
            return 45
        elif 4500000 <= x['last_price'] < 10000000:
            return 60
        else:
            return 90
    except:
        pass
data['exposition_price'] = data.apply(days_exposition_price, axis=1)
```
